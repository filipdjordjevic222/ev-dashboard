```python
def load_sales_data():
    sales_url = "https://api.iea.org/evs?parameters=EVsales&category=Historical&mode=Cars&csv=true"
    ev_sales_df = pd.read_csv(sales_url)
    return ev_sales_df

def load_charging_points_data():
    charging_url = "https://api.iea.org/evs?parameters=EV charging points&category=Historical&mode=EV&csv=true"
    ev_charging_points_df = pd.read_csv(charging_url)
    return ev_charging_points_df

def load_data():
    ev_sales_df = load_sales_data()
    ev_charging_points_df = load_charging_points_data()
    return ev_sales_df, ev_charging_points_df
```
